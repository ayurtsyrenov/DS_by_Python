```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils

```

```python
df = pd.read_csv('monthly-car-sales-in-quebec-1960.csv', sep=';', header=0, parse_dates=[0])
df.head(13)


```

```python
df.iloc[:,1].plot()
```

```python
# преобразуем данные
# надо построить авторегрессию из временного ряда
# X: 1 ... 12 Y:13
# X1: 2,..., 13 Y:14

df2 = pd.DataFrame()

for i in range(12,0,-1):
    df2['t-'+str(i)] = df.iloc[:,1].shift(i)
    
df2['t'] = df.iloc[:,1].values

print(df2.head(13))

```

```python
# избавимся от NAN
# cut first 12 rows

df3 = df2[12:]
df3.head()

```

```python
y = df3['t']

X = df3.drop('t', axis=1)
```

```python
# делим выборку на тренировочную и тестовую

X_train = X[:91]
y_train = y[:91]
X_test = X[91:]
y_test = y[91:]
```

```python
print(df3.shape)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# преобразуем из pandas to numpy

X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
# обучаем нейронную сеть

model = Sequential()
model.add(Dense(8, input_dim=12, activation='relu'))
model.add(Dense(1, activation='linear'))

# compiling model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_percentage_error'])

# Training of model
model.fit(X_train, y_train, epochs = 300, batch_size=None)


```

```python
# оценка качества на тестовой выборке

scores = model.evaluate(X_test, y_test)
print("\nMAPE: %.2f%%" % (scores[1]))
```

```python
# prediction

predictions = model.predict(X_test)
```

```python
predictions_train = model.predict(X_train)
```
