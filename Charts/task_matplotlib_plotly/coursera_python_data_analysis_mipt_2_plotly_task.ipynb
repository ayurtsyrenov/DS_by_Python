```
df = pd.read_csv('/content/CompleteDataset.csv')
```

```
# Смотрим в каком формате заданы стоимости игроков
df['Value'].head(10)
```

```
# Создаем новый столбец с численными значениями стоимостей
# Убираем из стоимости игроков символ '€'
# Убираем из стоимости игроков символ 'M' и умножаем на 1000000, убираем символ 'K' и умножаем на 1000
df['value_cleaned'] = df['Value'].str.strip('€').apply(lambda x: float(x.strip('M')) * 1000000 if 'M' in x else float(x.strip('K')) * 1000)
```

```
# Смотрим в каком формате заданы зарплаты игроков
df['Wage'].head()
```

```
# Создаем новый столбец с численными значениями запрплат
# Убираем из зарплат игроков символ '€'
# Убираем из зарплат игроков символ 'K' и умножаем на 1000
df['wage_cleaned'] = df['Wage'].str.strip('K€').apply(lambda x: int(x)*1000)
```

```
# Создаем датафреймы с игроками топовых клубов FC Bayern Munich, Real Madrid CF, FC Barcelona, Juventus
bayer = df[df['Club']=='FC Bayern Munich']
real = df[df['Club']=='Real Madrid CF']
barca = df[df['Club']=='FC Barcelona']
juventus = df[df['Club']=='Juventus']
```

```
# Рисуем график зависимости стоимости игроков от их Perfomance
# Видно, что начиная с Perfomance > 80 начинается значительный рост стоимости игроков

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=bayer['Overall'],
    y=bayer['value_cleaned'],
    name="FC Bayern Munich",
    mode='markers',
    text=bayer['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=barca['Overall'],
    y=barca['value_cleaned'],
    name="FC Barcelona",
    mode='markers',
    text=barca['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=real['Overall'],
    y=real['value_cleaned'],
    name="Real Madrid CF",
    mode='markers',
    text=real['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=juventus['Overall'],
    y=juventus['value_cleaned'],
    name="Juventus",
    mode='markers',
    text=juventus['Name'],
    marker = {'size': 10}
))

fig.update_layout(
    title="How players' value depends on their perfomance",
    xaxis_title="Player's overall perfomance",
    yaxis_title="Player's value (€)",
    font=dict(
        family="Courier New, monospace",
        size=14,
        color="#7f7f7f"
    )
)

fig.show()
```

```
# Рисуем график зависимости зарплат игроков от их Perfomance
# Видно, что начиная с Perfomance > 85 начинается экспоненциальный рост зарплат игроков

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=bayer['Overall'],
    y=bayer['wage_cleaned'],
    name="FC Bayern Munich",
    mode='markers',
    text=bayer['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=barca['Overall'],
    y=barca['wage_cleaned'],
    name="FC Barcelona",
    mode='markers',
    text=barca['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=real['Overall'],
    y=real['wage_cleaned'],
    name="Real Madrid CF",
    mode='markers',
    text=real['Name'],
    marker = {'size': 10}
))

fig.add_trace(go.Scatter(
    x=juventus['Overall'],
    y=juventus['wage_cleaned'],
    name="Juventus",
    mode='markers',
    text=juventus['Name'],
    marker = {'size': 10}
))

fig.update_layout(
    title="How players' wages depend on their perfomance",
    xaxis_title="Player's overall perfomance",
    yaxis_title="Player's wage (€)",
    font=dict(
        family="Courier New, monospace",
        size=14,
        color="#7f7f7f"
    )
)

fig.show()
```
