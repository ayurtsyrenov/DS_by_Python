```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('CompleteDataset.csv')
```

```python
# для удобства работы, добавим 
#соответствующие столбцы датасета в переменные

age = df.Age
skills = df.Overall
nation = df.Nationality
price = df.Value
```

```python
# Гисторгамма распределения переменной age
# покажет самый распространенный возраст среди существующих профессиональных футболистов
plt.hist(age)

plt.title('Гистограмма распределения возраста', fontdict={'fontsize':15})
plt.xlabel('Возраст')
plt.ylabel('Количество игроков');

# можно заметить, что распределение близко к нормальному
# в среднем большая часть футболистов в возрасте 20-28 лет
```

```python
data = np.column_stack((age, skills))

fig, (ax1,ax2) = plt.subplots(
    nrows=1, ncols = 2, figsize=(8,4))

ax1.scatter(x=age, y=skills, marker='o',
           c='r', edgecolor='b')
ax1.set_title('Распределение навыков по возрасту')
ax1.set_xlabel('Возраст')
ax1.set_ylabel('Навык')

ax2.hist(
    data, bins=np.arange(data.min(), data.max()),
    label=('Возраст', 'Навык'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Hist')
ax2.yaxis.tick_right()

plt.show()
```

```python
overall=df.Overall
pot=df.Potential


fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))


ax1.hist(overall, bins = np.arange(overall.min(), overall.max()), color = 'red', alpha = 1, label = 'overall')
ax1.hist(pot, bins = np.arange(pot.min(), pot.max()), color = 'green', alpha = 0.75, label = 'potential')
ax1.set_ylabel('count')
ax1.set_xlabel('value')
ax1.set_title('Name')
ax1.legend(loc = 'best')



plt.show()
```

```python
# Интерактивная визуадизация с помощью Plotly

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objects as go

import plotly.offline as offline

count_age_fifa = df.groupby('Age', as_index = False).ID.count()

trace = go.Bar(
    x = count_age_fifa.Age,
    y = count_age_fifa.ID
)
layout = go.Layout(
    title='Age of players',
)

fig = go.Figure(data = [trace], layout = layout)
iplot(fig)
```
