```python
import math
from sklearn import preprocessing

# стандартизируем данные

def scale_features(df1):
    scaled = preprocessing.StandardScaler().fit_transform(df1)
    scaled = pd.DataFrame(scaled, columns=df1.columns)
    return scaled.head()
    
```

```python
df1 = df
scale_features(df1)
```

```python
from sklearn.decomposition import FactorAnalysis

df_scaled = preprocessing.scale(df)
fa = FactorAnalysis(n_components=2)
fa.fit(df_scaled)
```

```python
pd.DataFrame(fa.components_, columns=df.columns)
```

```python
fa.noise_variance_
```

```python
print('Uniqueness:')
pd.Series(fa.noise_variance_, df.columns)
```

```python
scores = pd.DataFrame(fa.transform(df_scaled), columns=['factor1', 'factor2'])
scores

# можно сделать вывод, что степень развития процесса
# перехода к промышленному производству у 2 кантона очень нихок
# в то время, как успеха в этом деле добивается 18 и 19 кантоны
# лидер по промышленной революции de_geneva(45 в списке исходных данных)

```
