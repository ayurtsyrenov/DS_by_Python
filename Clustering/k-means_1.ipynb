```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,8)
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2,
                 centers=4, random_state=1)
```

```python
plt.scatter(X[:, 0], X[:,1]);
```

```python
from sklearn.cluster import KMeans
```

```python
# зададим параметры метода к-средних

kmeans = KMeans(n_clusters=2, init='k-means++',
                  n_init=10, random_state=1)

```

```python
# приступим к обучению модели функциоей fit

kmeans.fit(X)

```

```python
labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:,1], c=labels);
```

```python
# подберем наилучшее число К (количество кластеров)
# прибегнем к методу к-средних для разных количества кластеров
# данный метод - метод Локтя
# будем смотреть на значения к-средних, предварительно сохраним в список

crit = []

for k in range(2,8):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    
    crit.append(kmeans.inertia_)
```

```python
plt.plot(range(2,8), crit)

plt.title('График локтя')
plt.xlabel('k')
plt.ylabel('k-means');

# по критерию локтя количество кластеров = 4
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++',
                  n_init=10, random_state=1)

```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:,1], c=labels);
```
